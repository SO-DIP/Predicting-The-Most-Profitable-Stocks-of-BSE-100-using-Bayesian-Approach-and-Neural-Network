```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from sklearn.preprocessing import MinMaxScaler
from statsmodels.tsa.stattools import adfuller
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout


# Download the data
df = pd.read_csv("Dabur.csv" , index_col = 'Date', parse_dates = ['Date'])
df =df[::-1]
```

```python
df
```

```python
df = df['Close Price']
```

```python
from scipy.stats import shapiro
data =  df
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f\n' % (stat, p))
if p > 0.05:
    print('probably Gaussian')
else:
    print('probably not Gaussian')
```

```python
adf_result = adfuller(df)
if adf_result[1] > 0.05:
    print('The data is not stationary. Detrending the data...')
```

```python
from scipy import signal
df = signal.detrend(df)
df
```

```python
# Preprocess the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(np.array(df).reshape(-1, 1))

# Split data into training and testing sets
training_size = int(len(scaled_data) * 0.8)
test_size = len(scaled_data) - training_size
train_data, test_data = scaled_data[0:training_size,:], scaled_data[training_size:len(scaled_data),:]

# Define the look-back period
look_back = 60

# Define function to create dataset with look-back period
def create_dataset(dataset, look_back=60):
    X, Y = [], []
    for i in range(look_back, len(dataset)):
        X.append(dataset[i-look_back:i, 0])
        Y.append(dataset[i, 0])
    return np.array(X), np.array(Y)

# Create training and testing datasets
X_train, y_train = create_dataset(train_data, look_back)
X_test, y_test = create_dataset(test_data, look_back)
```

```python
scaled_data
```

```python
# Reshape input data to be suitable for LSTM
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
# Define LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1))

# Compile and fit model
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=100, batch_size=32)

# Make predictions on test set
y_pred = model.predict(X_test)

# Invert predictions and actual values from scaling
y_pred = scaler.inverse_transform(y_pred)
y_test = scaler.inverse_transform([y_test])


# Plot actual and predicted values
plt.figure(figsize=(12,8))
plt.plot(y_test.flatten(), label="Actual")
plt.plot(y_pred.flatten(), label="Predicted")
plt.legend(fontsize=12)
plt.xlabel("Days", fontsize=12)
plt.ylabel("Closing Price", fontsize=12)
plt.title("Dabur India Stock Price Prediction", fontsize=16)
plt.show()
```
